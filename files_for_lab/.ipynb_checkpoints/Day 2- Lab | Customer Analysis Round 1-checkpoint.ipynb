```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('csv_files/file1.csv')
df1.head(5)
```

```python
df2 = pd.read_csv('csv_files/file2.csv')
df2.head(3)
```

```python
df3 = pd.read_csv('csv_files/file3.csv')
df3.head(3)
```

```python
df1.shape, df2.shape, df3.shape
```

```python
df1 = df1.rename(columns={'GENDER':'Gender', 'ST':'State'})
df2 = df2.rename(columns={'GENDER':'Gender', 'ST':'State'})
```

```python
column_names = df1.columns
df = pd.DataFrame(columns=column_names)
df = pd.concat([df,df1,df2,df3],axis= 0)
df.shape
```

```python
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower())
df.columns = cols
df.head()
```

```python
df = df[['customer','state','gender','customer lifetime value','income','monthly premium auto','policy type','vehicle class','total claim amount']]
```

```python
df.head()
```

```python
df.columns = [i.replace(' ','_') for i in df.columns]
```

```python
df.info()
```

```python
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
```

```python
df['customer_lifetime_value'].dtype
```

```python
df['customer_lifetime_value']  = df['customer_lifetime_value'].fillna(0)
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(int)
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'] * 100
```

```python
df.tail()
```

```python
df.drop_duplicates(inplace = True)
```
